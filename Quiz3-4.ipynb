```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Earthquakes_database.csv")
```

# 1. Data Cleaning

```python
#split date
df[["Month", "Day","Year"]] = df['Date'].str.extract('(\d+)/(\d+)/(\d+)')
df[["Month", "Day","Year"]]
```

```python
#split ID to Text and Numbers and add new columns
df[["ID Text", "ID numbers"]] = df["ID"].str.extract('([a-zA-Z]+)([0-9]+)')
df[["ID Text", "ID numbers"]]
```

```python
#drop ID column
df = df.drop(columns="ID")
```

```python
df
```

```python
#fill NaN values with N/A
df = df.fillna('N/A')
```

```python
df
```

```python
#change status automatic to Au, Reviewed to Rw
df["Status"] = df["Status"].str.replace("Automatic","Au")
df["Status"] = df["Status"].str.replace("Reviewed","Rw")
df["Status"]
```

# 2. Create categories

```python
#create categories for magnitude and add new column
magnitude_bins = [df['Magnitude'].min() - 1, 5.0, 6.0, df['Magnitude'].max() + 1]
magnitude_labels = ['Low', 'Medium', 'High']

df['magnitude_categories'] = pd.cut(df['Magnitude'], bins=magnitude_bins, labels=magnitude_labels)
```

```python
df['magnitude_categories']
```

# 3 Data Reshaping

```python
#Pivot
df.pivot(columns="Year")
```

```python
#Create a pivot table
pivot_table_result = pd.pivot_table(df, values='Magnitude', index='Magnitude Type', columns='Type', aggfunc='mean')

# create a crosstab
crosstab_result = pd.crosstab(index=df['Magnitude Type'], columns=df['Type'], values=df['Magnitude'], aggfunc='mean')

print("Pivot Table:")
print(pivot_table_result)

print("\nCrosstab:")
print(crosstab_result)
```

# 4. Charts

```python
# Line Chart of magnitude mean by year
fig, ax = plt.subplots(figsize=(15, 6))

plt.plot(df.groupby('Year')['Magnitude'].mean().head(25), marker='o', linestyle='-', color='green', label='Average Magnitude')

plt.xlabel('Year')
plt.ylabel('Average Magnitude')
plt.title('Line Chart of Average Magnitudes by Year')
plt.legend()

# Show the plot
plt.grid(True)  
plt.show()
```

```python
# Bar Chart for event types 
type_counts = df['Type'].value_counts()
type_counts.plot(kind='bar', color='skyblue', edgecolor='black', title='Number of Events by Type')
plt.xlabel('Event Type')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter Chart for relationshio between magnitude and depth
df.plot(x='Magnitude', y='Depth', kind='scatter', title='Magnitude / Depth',  marker='*', color='red')
plt.xlabel('Magnitude')
plt.ylabel('Depth')
plt.show()
```

```python
# box plot for magnitudes
fig, ax = plt.subplots(figsize=(8, 6))

ax.boxplot(df['Magnitude'], vert=False, patch_artist=True, labels=['Magnitude'], boxprops=dict(facecolor='lightblue'))

ax.set_title('Box Plot of Magnitudes')

plt.show()
```

# 5. Date Functions

```python
#date functions
d_parser = lambda x: pd.to_datetime(x, format='%m/%d/%Y', errors='coerce')

df2 = pd.read_csv('Earthquakes_database.csv', parse_dates=['Date'], date_parser=d_parser)
df2
```

```python
#add day of week for data
df2['DayOfWeek'] = df2['Date'].dt.day_name()
```

```python
df2['DayOfWeek']
```

```python
# Resample the data to a weekly frequency, calculating the mean for 'Magnitude' and maximum for 'Depth'
df2['Date'] = pd.to_datetime(df2['Date'], errors='coerce')

if df2['Date'].dt.tz is None:
    df2['Date'] = df2['Date'].dt.tz_localize('UTC')

df2.set_index('Date', inplace=True)

resampled_df = df2.resample('W').agg({'Magnitude': 'mean', 'Depth': 'max'})

resampled_df
```
